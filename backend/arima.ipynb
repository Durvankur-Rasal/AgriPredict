```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.stattools import adfuller
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from sklearn.metrics import mean_squared_error
from math import sqrt
import warnings
warnings.filterwarnings('ignore')

# Function to load and prepare the data
def load_data(filepath):
    """
    Load crop price data from CSV file and prepare it for time series analysis
    """
    # Load data
    df = pd.read_csv(filepath)
    
    # Convert Price Date to datetime format
    df['Price Date'] = pd.to_datetime(df['Price Date'], dayfirst=True)
    
    # Sort by date
    df = df.sort_values('Price Date')
    
    return df

# Function to analyze data for a specific crop and market
def analyze_crop_data(df, crop_name, market_name=None):
    """
    Filter data for a specific crop and market, then prepare it for time series analysis
    """
    # Filter data
    if market_name:
        filtered_df = df[(df['Commodity'] == crop_name) & (df['Market Name'] == market_name)]
    else:
        filtered_df = df[df['Commodity'] == crop_name]
    
    # Resample data to monthly frequency and calculate average prices
    # Group by date and calculate average price
    time_series = filtered_df.groupby('Price Date')['Modal Price (Rs./Quintal)'].mean()  #create time series
    
    # Convert to monthly frequency if needed
    if len(time_series) > 12:  # If we have enough data points
        time_series = time_series.resample('MS').mean().fillna(method='ffill')
    
    return time_series

# Function to check stationarity with Augmented Dickey-Fuller test
def check_stationarity(time_series):
    """
    Check if the time series is stationary using the ADF test
    """
    if len(time_series) < 5:
        print("Warning: Time series too short for reliable stationarity testing.")
        return False
        
    result = adfuller(time_series.dropna())
    print(f'ADF Statistic: {result[0]}')
    print(f'p-value: {result[1]}')
    print('Critical Values:')
    for key, value in result[4].items():
        print(f'\t{key}: {value}')
    
    # Interpret the results
    if result[1] <= 0.05:
        print("Conclusion: The series is stationary")
        return True
    else:
        print("Conclusion: The series is not stationary")
        return False

# Function to make the time series stationary through differencing
def make_stationary(time_series, d=1):
    """
    Apply differencing to make the time series stationary
    """
    return time_series.diff(d).dropna()


# Function to fit ARIMA model and forecast
def fit_arima_and_forecast(time_series, p, d, q, forecast_steps=12):
    """
    Fit ARIMA model and forecast future prices
    """
    # Check if we have enough data
    if len(time_series) < p + d + q + 5:
        print(f"Warning: Time series too short for ARIMA({p},{d},{q}). Need at least {p+d+q+5} points, but have {len(time_series)}.")
        print("Trying a simpler model...")
        p = min(p, 1)
        q = min(q, 1)
    
    # Split data into train and test sets (if enough data)
    if len(time_series) >= 5:
        train_size = int(len(time_series) * 0.8)
        train_size = max(train_size, 3)  # At least 3 points for training
        test_size = len(time_series) - train_size
        
        if test_size > 0:
            train, test = time_series[:train_size], time_series[train_size:]
        else:
            train, test = time_series, None
    else:
        train = time_series
        test = None
    
    # Fit ARIMA model
    try:
        model = ARIMA(train, order=(p, d, q))
        model_fit = model.fit()
        
        # Print model summary
        print(model_fit.summary())
        
        # In-sample predictions
        in_sample_pred = model_fit.predict(start=0, end=len(train)-1)
        
        # Out-of-sample forecast
        if test is not None and len(test) > 0:
            forecast = model_fit.forecast(steps=len(test))
            
            # Calculate RMSE
            rmse = sqrt(mean_squared_error(test, forecast))
            print(f"RMSE: {rmse}")
        else:
            forecast = None
        
        # Plot results
        plt.figure(figsize=(12, 6))
        plt.plot(time_series.index, time_series, label='Actual')
        plt.plot(in_sample_pred.index, in_sample_pred, label='Fitted', color='red')
        
        if forecast is not None and len(forecast) > 0:
            plt.plot(test.index, forecast, label='Forecast', color='green')
        
        plt.title(f'ARIMA({p},{d},{q}) Model: Actual vs Forecast')
        plt.legend()
        plt.grid(True)
        plt.show()
        
        # Future forecast
        future_forecast = model_fit.forecast(steps=forecast_steps)
        
        # Generate dates for forecasted values
        last_date = time_series.index[-1]
        if isinstance(last_date, pd.Timestamp):
            # If we have a timestamp index
            freq = 'MS' if len(time_series) > 12 else 'D'  # Monthly or daily frequency
            future_dates = pd.date_range(start=last_date, periods=forecast_steps+1, freq=freq)[1:]
        else:
            # If we have a numeric index
            future_dates = range(len(time_series), len(time_series) + forecast_steps)
        
        plt.figure(figsize=(12, 6))
        plt.plot(time_series.index, time_series, label='Historical')
        plt.plot(future_dates, future_forecast, label='Future Forecast', color='green')
        plt.title(f'ARIMA({p},{d},{q}) Model: Future Price Forecast')
        plt.legend()
        plt.grid(True)
        plt.show()
        
        return model_fit, future_forecast, future_dates
    
    except Exception as e:
        print(f"Error fitting ARIMA model: {e}")
        print("Try using different parameters or check your data.")
        return None, None, None

# Function to analyze seasonal patterns
def analyze_seasonal_patterns(time_series, crop_name):
    """
    Analyze and visualize seasonal patterns in crop prices
    """
    if len(time_series) < 12:
        print(f"Warning: Not enough data points to analyze seasonal patterns for {crop_name}.")
        return None
    
    # Extract month from datetime index
    if isinstance(time_series.index[0], pd.Timestamp):
        time_series_month = time_series.copy()
        time_series_month.index = time_series_month.index.month
        
        # Group by month and calculate average price
        monthly_avg = time_series_month.groupby(time_series_month.index).mean()
        
        # Plot seasonal patterns
        plt.figure(figsize=(12, 6))
        monthly_avg.plot(kind='bar', color='skyblue')
        plt.title(f'Seasonal Price Pattern for {crop_name}')
        plt.xlabel('Month')
        plt.ylabel('Average Price (Rs./Quintal)')
        plt.xticks(range(1, 13), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
        plt.grid(True, axis='y')
        plt.show()
        
        return monthly_avg
    else:
        print("Warning: Time series index is not datetime format. Cannot analyze seasonal patterns.")
        return None

# Function to compare crops across seasons
def compare_crops_seasonal(df, crop_list):
    """
    Compare seasonal price patterns across different crops
    """
    plt.figure(figsize=(14, 8))
    
    for crop in crop_list:
        # Get time series for the crop
        crop_ts = analyze_crop_data(df, crop)
        
        if len(crop_ts) < 12:
            print(f"Warning: Not enough data points for seasonal analysis of {crop}.")
            continue
            
        if not isinstance(crop_ts.index[0], pd.Timestamp):
            print(f"Warning: Index for {crop} is not in datetime format.")
            continue
        
        # Extract month and calculate average price by month
        crop_ts_month = crop_ts.copy()
        crop_ts_month.index = crop_ts_month.index.month
        monthly_avg = crop_ts_month.groupby(crop_ts_month.index).mean()
        
        # Check if we have enough months
        if len(monthly_avg) < 6:
            print(f"Warning: Not enough months in data for {crop}.")
            continue
        
        # Normalize prices for better comparison
        if monthly_avg.max() != monthly_avg.min():
            normalized_prices = (monthly_avg - monthly_avg.min()) / (monthly_avg.max() - monthly_avg.min())
            
            # Plot normalized seasonal pattern
            plt.plot(normalized_prices.index, normalized_prices, label=crop, marker='o')
    
    plt.title('Normalized Seasonal Price Patterns - Crop Comparison')
    plt.xlabel('Month')
    plt.ylabel('Normalized Price')
    plt.xticks(range(1, 13), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
    plt.legend()
    plt.grid(True)
    plt.show()

# Function to find optimal ARIMA parameters using AIC
def find_optimal_arima_params(time_series, max_p=3, max_d=2, max_q=3):
    """
    Find optimal ARIMA parameters using AIC
    """
    if len(time_series) < 10:
        print("Warning: Time series too short for parameter optimization.")
        return 1, 0, 1
        
    best_aic = float('inf')
    best_params = None
    
    for p in range(max_p + 1):
        for d in range(max_d + 1):
            for q in range(max_q + 1):
                if p == 0 and q == 0:
                    continue  # Skip ARIMA(0,d,0) as it's just differencing
                    
                try:
                    model = ARIMA(time_series, order=(p, d, q))
                    model_fit = model.fit()
                    aic = model_fit.aic
                    
                    if aic < best_aic:
                        best_aic = aic
                        best_params = (p, d, q)
                        
                    print(f"ARIMA({p},{d},{q}) - AIC: {aic}")
                except:
                    continue
    
    if best_params is None:
        print("Could not find optimal parameters. Using default (1,0,1).")
        return 1, 0, 1
    
    print(f"Best ARIMA parameters: {best_params} with AIC: {best_aic}")
    return best_params

# Main function to run the analysis
def main():
    """
    Main function to run the ARIMA analysis for crop price prediction
    """
    # Load data
    filepath = 'historical_prices.csv'  # Replace with your actual file path
    try:
        df = load_data(filepath)
        print("Data loaded successfully!")
        print(f"Dataset shape: {df.shape}")
        print("First few rows of the dataset:")
        print(df.head())
    except Exception as e:
        print(f"Error loading data: {e}")
        return
    
    # Get all unique crops in the dataset
    crops = df['Commodity'].unique()
    print(f"\nUnique crops in the dataset: {crops}")
    
    # For each crop, perform time series analysis
    for crop in crops:
        print(f"\n{'='*50}")
        print(f"Analysis for {crop}")
        print(f"{'='*50}")
        
        # Get time series for the crop
        ts = analyze_crop_data(df, crop)
        
        if len(ts) < 5:
            print(f"Warning: Not enough data points for {crop}. Skipping detailed analysis.")
            continue
            
        print(f"\nTime series for {crop}:")
        print(ts.head())
        print(f"Number of data points: {len(ts)}")
        
        # Plot the time series
        plt.figure(figsize=(12, 6))
        ts.plot()
        plt.title(f'{crop} Price Trend')
        plt.xlabel('Date')
        plt.ylabel('Price (Rs./Quintal)')
        plt.grid(True)
        plt.show()
        
        # Check stationarity
        print("\nChecking stationarity:")
        is_stationary = check_stationarity(ts)
        
        # If not stationary, apply differencing
        d_value = 0
        diff_ts = ts
        
        if not is_stationary and len(ts) >= 5:
            print("\nApplying first-order differencing to make the series stationary...")
            diff_ts = make_stationary(ts)
            d_value = 1
            
            # Check stationarity again
            if len(diff_ts) >= 5:
                is_stationary_diff = check_stationarity(diff_ts)
                
                if not is_stationary_diff:
                    print("\nApplying second-order differencing...")
                    diff2_ts = make_stationary(diff_ts)
                    d_value = 2
                    
                    if len(diff2_ts) >= 5:
                        check_stationarity(diff2_ts)
                        diff_ts = diff2_ts
        
        # Find optimal ARIMA parameters
        print("\nFinding optimal ARIMA parameters...")
        if len(diff_ts) >= 10:
            p, d, q = find_optimal_arima_params(ts, max_p=2, max_d=d_value, max_q=2)
        else:
            print("Not enough data for parameter optimization. Using default values.")
            p, q = 1, 1
            d = d_value
        
        # Fit ARIMA model and forecast
        print(f"\nFitting ARIMA({p},{d},{q}) model...")
        model, forecast, forecast_dates = fit_arima_and_forecast(ts, p, d, q)
        
        # Analyze seasonal patterns if we have enough data
        if len(ts) >= 12:
            print("\nAnalyzing seasonal patterns...")
            monthly_avg = analyze_seasonal_patterns(ts, crop)
            
            if monthly_avg is not None and len(monthly_avg) > 0:
                best_month = monthly_avg.idxmax()
                month_name = {1: 'January', 2: 'February', 3: 'March', 4: 'April', 5: 'May', 6: 'June', 
                             7: 'July', 8: 'August', 9: 'September', 10: 'October', 11: 'November', 12: 'December'}
                print(f"Best month to sell {crop}: {month_name.get(best_month, best_month)} with average price of {monthly_avg.max():.2f} Rs./Quintal")
        
        # Print forecast results
        if forecast is not None and len(forecast) > 0:
            print("\nForecast for next periods:")
            for date, price in zip(forecast_dates, forecast):
                print(f"{date}: {price:.2f} Rs./Quintal")
    
    # Compare seasonal patterns across crops if we have multiple crops
    if len(crops) > 1:
        print("\nComparing seasonal patterns across crops...")
        compare_crops_seasonal(df, crops)

if __name__ == "__main__":
    main()
```
